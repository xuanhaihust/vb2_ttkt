```python
#library
import tensorflow as tf
from tensorflow.keras.applications import mobilenet
from tensorflow.keras import layers
from tensorflow.keras.models import Model
import tensorflow.keras.backend as K
from tensorflow.keras.callbacks import Callback, History
import numpy as np
import os, pickle
from PIL import Image

import matplotlib.pyplot as plt
```

```python
#use mobilenet model for quicly
def build_model():
    model = mobilenet.MobileNet(include_top = False, input_shape = (160, 160, 3), weights = "imagenet"
                                , pooling="avg")
    x = layers.Dense(2)(model.layers[-1].output)
    x = layers.Lambda(multitask_activation)(x)
    return Model(model.layers[0].input, x)
```

```python
#define multitask activation
def multitask_activation(inputs):
    sigmoid = K.sigmoid(inputs[:, 0])
    relu = K.relu(inputs[:, 1])
    result = K.stack([sigmoid, relu], axis=-1)
    return result
```

```python
#predict
def predict(img_path):
    model = build_model()
    model.load_weights("./mobileNet.hdf5")
    #pre-process image
    img = np.asarray(Image.open(img_path).convert("RGB").resize((160, 160)), np.uint8 )
    feature = mobilenet.preprocess_input(np.expand_dims(img, axis = 0))
    
    #predict
    pred = model.predict(feature)
    
    prob_male , age = pred[0]
    gender_default = "Male"
    prob = prob_male
    if prob_male < 0.5:
        gender_default = "Female"
        prob = 1 - prob_male
    print(img_path, "Gender :", gender_default, "percent: {:.2%}".format(prob), " | Age : %2.0f" %age)
```

```python
path = "050742.jpg"
img = Image.open(path)
plt.imshow(img)

#predict
predict(path)
```
